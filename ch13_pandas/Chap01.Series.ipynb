<h5><font color = "BlueViolet">시리즈의 값 읽어오기</font></h5>

```python
myindex02 = ['아메리카노','에스프레소','카페라떼','카페모카','콜드브루','바닐라라떼','마끼아또']
mylist02 = [50,30,20,60,40,80,70]
myseries05 = pd.Series(data=mylist02, index=myindex02)
print(myseries05)
```

```python
myseries05['카페모카'] # scalar로 변환
```

```python
myseries05[['카페모카']] # Series로 반환
```

```python
myseries05['에스프레소':'카페모카'] # 문자열 라벨링 : from <= 범위 <= to
```

```python
myseries05[['마끼아또','카페모카']] # 콤마 기호는 서로 떨어진 데이터 추출
```

```python
myseries05.iloc[3] # i(index) loc(location) 정수 위치 기반 접근 방식, 문자는 loc를 이용해서 문자 기반 접근을 한다.
```

```python
myseries05.iloc[0:7:3]
```

```python
myseries05.iloc[3:6]
```

<h5><font color = "BlueViolet">시리즈에 값 쓰기</font></h5>

```python
myseries05.iloc[2] = 2000
myseries05.iloc[2:5] = 5000
print(myseries05)
```

```python
myseries05.loc[['아메리카노','카페모카']] = 3000
print(myseries05)
```

```python
myseries05[0::2] = 7000
print(myseries05)
```

```python
result = myseries05[myseries05 <= 4000] # 연산자로 필터링하기
print(result)
```

<h5 style="color: red;">시리즈 연산</h5>

<h6><font color = "BlueViolet">시리즈 샘플 만들기</font></h6>

```python
myindex31 = ['에스프레소','아메리카노','마키아또', '바닐라라떼']
mylist31 = [30,40,50,60]

myindex32 = ['에스프레소','아메리카노','마키아또', '콜드브루']
mylist32 = [20,40,60,70]
myseries31 = pd.Series(data=mylist31,index=myindex31)
myseries32 = pd.Series(data=mylist32,index=myindex32)
```

```python
myseries31
```

```python
myseries32
```

```python
su = 10
print(myseries31 + su)
```

```python
print(myseries31.add(su))
```

```python
print(myseries31.sub(su))
```

```python
print(myseries31.mul(su))
```

```python
print(myseries31.div(su)) # floordiv() 함수도 있음
```

<h6><font color = "BlueViolet">관계 연산자와 관련된 메소드</font></h6>

```python
myseries31[myseries31 >=50]
```

```python
myseries31[myseries31.ge(50)]
```

```python
myseries31[myseries31.gt(50)]
```

```python
myseries31[myseries31.eq(50)]
```

```python
newseries = myseries31 + myseries32 # NaN : Not a Number
print(newseries)
```

```python
newseries = myseries31.sub(myseries32, fill_value=0)
print(newseries)
```

```python
newseries = myseries31.mul(myseries32, fill_value=12)
print(newseries)
```

<h6><font color = "BlueViolet">unique 메소드 사용</font></h6>

```python
mylist04 = ['카페모카','에스프레소','에스프레소','카페라떼','에스프레소','에스프레소','아메리카노','아메리카노','카페모카','카페모카']
myseries04 = pd.Series(mylist04)
myseries04
```

```python
myunique04 = myseries04.unique()
myunique04
```

```python
myseries04.value_counts() # 빈도수 파악해서 역순으로 정렬
```

```python
myseries04.value_counts(normalize= True) # 빈도수를 비율로 보여주기
```

```python
mask04 = myseries04.isin(['카페모카','카페라떼'])
mask04
```

```python
myseries04[mask04]
```

<h5><font color = "red">그래프 그리기 </font></h5>

```python
myindex05 = ['아메리카노','에스프레소','카페라떼','카페모카','콜드브루','바닐라라떼', '마끼아또']
mylist05 = [5000,3000,2000,6000,4000,8000,7000]
myseries05 = pd.Series(data=mylist05,index=myindex05)
myseries05
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic' # `한글 글꼴 깨짐 방지
```

```python
myseries05.plot(kind = 'line', grid = True, rot = 20)
plt.xlabel('상품명')
plt.ylabel('단가')
plt.title('품목별 단가 꺾은선', fontsize = 18)
```

```python
myseries05.plot(kind = 'bar', grid = True, rot = 0, color = ['r','g','b','y','m'])
plt.xlabel('상품명')
plt.ylabel('단가')
plt.title('품목별 단가 꺾은선', fontsize = 18)
```
