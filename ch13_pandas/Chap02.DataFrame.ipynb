```python
condition01 = (myframe02['스타벅스'] >= 70)
condition01
```

```python
condition02 = (myframe02['메가커피'] >= 140)
condition02
```

```python
# and, or, not 연산자는 일대일 비교라서 사용하지 못합니다.
# 대신 element wise(원소별 비교)를 위하여 &(and), |(or), ~(not)을 사용합니다
boolResult = condition01 & condition02
result02 = myframe02.loc[boolResult]
result02
```

```python
print('양쪽 결과 통합')
condition = pd.DataFrame([condition01,condition02])
condition.T
```

```python
condition.all() # and 조건과 비슷 (모든 조건에 참인것만)
```

```python
condition.any() # or 조건과 비슷 (하나라도 참이면)
```

```python
print('all() 함수를 사용하여 모든 조건에 부합하는 데이터 찾기')
result02 = myframe02.loc[condition.all()]
result02
```

```python
print('any() 함수를 사용하여 조건중에 하나라도 부합하는 데이터 찾기')
result02 = myframe02.loc[condition.any()]
result02
```

```python
myindex04 = ['아메리카노', '카페모카', '마끼아또']
mycolumns04 = ['투썸플레이스','스타벅스', '이디야커피']
mylist04 = [50,60,40,80,30,50,80,40,60]
data04 = np.reshape(mylist04, (3,3))
myframe04 = pd.DataFrame(data04, index=myindex04, columns= mycolumns04)
myframe04
```

```python
# 재색인 기능 : 색인을 추가/삭제/다시 정리하는 작업 (reindex)
print('새로운 품목 "바닐라라떼"가 입고되었습니다.')
newindex = ['아메리카노','카페모카','마끼아또','바닐라라떼'] # 신규 품목 : 바닐라라떼
myframe04 = myframe04.reindex(newindex, fill_value=50) # 입고 수량 : 50개
myframe04
```

```python
print('신규 브랜드 "빽다방"에 40개씩 입고되었습니다.')
newdata = ['투썸플레이스','스타벅스','이디야커피','빽다방']
myframe04 = myframe04.reindex(columns=newdata, fill_value=40)
myframe04
```

```python
print('카페모카는 더 이상 판매되지 않습니다.')
myframe04 = myframe04.drop(['카페모카'], axis=0) # axis = 0은 행 대상 작업
myframe04
```
